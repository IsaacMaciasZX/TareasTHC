```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
```

```python
def get_data(N=20):
  x=np.linspace(0,2,N)
  y=np.zeros_like(x)
  eps=1E-1
  for i in range(N):
   y[i]=2*x[i]+3+(eps*(-1)**i)

  return x,y
```

```python

def plot_data(x,y):
  z=2*x+3
  plt.plot(x,y,'r*',label="Samples") #samples, los asteriscos son estrellas
  plt.plot(x,z,'-k',label="Model") #Model
  plt.title('Linear Regression')
  plt.xlabel('X data')
  plt.ylabel('Y data')
  plt.grid(lw=2)
  plt.legend()
  plt.show()
x,y=get_data(40)
plot_data(x,y)
```

```python
def get_data(N=20):
  x=np.linspace(0,2,N)
  y=np.zeros_like(x)
  eps=1E-1
  for i in range(N):
   y[i]=3*x[i]**2+5*x[i]-2+5*(eps*(-1)**i)

  return x,y
```

```python
def plot_data(x,y,params): #Modelo de regresion cuadratica, se editaron las variables definidas (z en este caso)
  z=params[0]+params[1]*x+params[2]*x**2

  plt.figure()
  plt.plot(x,y,'r*',label="Samples") #samples, los asteriscos son estrellas
  plt.plot(x,z,'-k',label="Model") #Model
  plt.title('Cuadratic Regression')
  plt.xlabel('X data')
  plt.ylabel('Y data')
  plt.grid(lw=2)
  plt.legend()
  plt.show()
```

```python
def Cuadratic_regression(N=10):
  x,y=get_data(N)
  A=np.vstack([np.ones(len(x)),x,x**2]).T #se agregó una x^2
  #el b ahora es la y
  AtA=A.T@A
  Aty=A.T@y
  params=LA.solve(AtA,Aty)
  plot_data(x,y,params)
  Oridata=np.array([-2.,5.,3.])
  Error=LA.norm(params-Oridata)
  print("Error: ",Error)
  return params
```
